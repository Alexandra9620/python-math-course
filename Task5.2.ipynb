```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from pylab import *
from mpl_toolkits.mplot3d import Axes3D


fig = figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 1)
Y = np.arange(-10, 10, 1)
X, Y = np.meshgrid(X, Y)

# График 1 
A, B, C = 7, 5, 2
Z1 = X**2/A**2-Y**2/B**2
ax.plot_surface(X, Y, Z1)

# График 2
A, B, C = 3, 10, 2
Z2 = Z1 = X**2/A**2+Y**2/B**2
ax.plot_surface(X, Y, Z2)

show()
```
