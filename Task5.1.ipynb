```
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 5)
Y = np.arange(-5, 5, 5)
X, Y = np.meshgrid(X, Y)
Z2 = (-6 * X - 8 * Y - 8) / 4
Z1 = (-6 * X - 8 * Y) / 4
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
show()
```
