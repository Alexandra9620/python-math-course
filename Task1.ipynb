```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 2.0, 0.01)

k, a, b = 1, 2, 3
y1 = k * np.cos(x - a) + b

k, a, b = 4, 5, 6
y2 = k * np.cos(x - a) + b

fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2)

ax.set(xlabel='x', ylabel='y')

plt.show()
```
