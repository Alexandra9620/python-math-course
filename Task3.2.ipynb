```
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import math
```

```
x = []
y1 = []
y2 = []
a = 8
b = 4
for i in range(-100, 101):
    x_ = i / 100 * a
    y_ = i / 100 * b
    x.append(x_)
    y1.append(math.sqrt((1 - (x_ ** 2) / (a ** 2)) * (b ** 2)))
    y2.append(-math.sqrt((1 - (x_ ** 2) / (a ** 2)) * (b ** 2)))
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("x")
plt.ylabel("y")
```
